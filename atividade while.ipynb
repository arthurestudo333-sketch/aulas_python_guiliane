```python
#. Peça a idade e se a pessoa tem carteira de motorista mostre se ela pode dirigir.

idade_da_pessoa = int(input("Digite sua idade :"))
tem_documento = input ("Tem CNH? (s/n):")
if idade_da_pessoa >= 18 and tem_documento == "s":
    print ("Pode dirigir")
else :
    print ("Não pode dirigir")
```

```python
#. Peça um número. Verifique se ele está entre 50 e 100.

numerol = int(input("Digite um numero:"))

if numerol >=50 and numerol <=100:
    print("Numero dentro do intervalo")
else:
    print("Numero fora do intervalo")
```

```python
#Peça o dia da semana. Mostre se é final de semana.

dia = input("Digite o dia da semana:")

if dia == "sábado" or dia =="domingo":
    print("Final de semana")
else:
    print("Dia util")
```

```python
#Peça usuário e senha. Verifique se ambos estão corretos.

usuario = input("Digite seu usuario:")
senha = input("Digiete a senha:")

if usuario != "admin" or senha != "admin123":
    print("Acesso Negado")
else:
    print("Acesso Permitido")
```

```python
#. Peça uma nota e a frequência do aluno. Mostre se ele foi aprovado.

nota = input("Digite sua nota")
frequencia = input("Digite sua frequencia")

if nota >= "7" and frequencia >= "75":
    print("Aprovado")
else:
    print("Reprovado")
```

```python
#. Peça se o aluno entregou o trabalho. Use not para verificar se ele não entregou.

trabalho = input("Você entregou o trabalho? (s/n):")

if not trabalho == "s":
    print("Pendente")
else:
    print("Entregue")
```

```python
#1. Mostre os números de 1 até 10 usando while.

contador = 1

while contador <=10:
    print(contador)
    contador += 1  
```

```python
#2. Peça uma senha até o usuário acertar.

senha_correta = "1234"
senha = ""

while senha != senha_correta:
    senha = input("Digite a senha: ")
    
    if senha != senha_correta:
        print("Senha incorreta!")

print("Acesso liberado!")
```

```python
#3. Peça números até o usuário digitar 0.

numero = int(input("Digite um número: "))

while numero != 0:
    print("Incorreto")
    numero = int(input("Digite um número: "))

print("Correto!")
```

```python
#4. Conte de 10 até 1.

contador = 10

while contador >=1:
    print(contador)
    contador -= 1  
```

```python
#5. Peça uma idade válida até o usuário digitar um valor positivo.

idade = float(input("Digite sua idade: "))

while idade < 0:
    print("Idade inválida")
    idade = float(input("Digite sua idade: "))

print("Idade válida")
```

```python
#6. Crie um menu que só encerra quando o usuário digitar 0.

opcao = -1

while opcao != 0:
    print("\n=== MENU ===")
    print("1 - Opção 1")
    print("2 - Opção 2")
    print("3 - Opção 3")
    print("0 - Sair")
    
    opcao = int(input("Escolha uma opção: "))
    
    if opcao == 1:
        print("Você escolheu a opção 1")
    elif opcao == 2:
        print("Você escolheu a opção 2")
    elif opcao == 3:
        print("Você escolheu a opção 3")
    elif opcao == 0:
        print("Encerrando o programa...")
    else:
        print("Opção inválida!")
```
