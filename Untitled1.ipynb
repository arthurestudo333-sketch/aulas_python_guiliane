7. Classificação de idade

```python
idade=int(input("Digite sua idade:"))

if idade <13:
    print("Criança")

elif idade <18:
    print("Adolescente")

else:
    print("Adulto")
```

8. Nota do aluno

```python
provas=float(input("Digite a nota:"))

if provas >=7:
    print("Aprovada")
elif provas >=5:
    print("Recuperação")
else:
    print("Reprovado")
```

9. Comparar três números

```python
c1=float(input("Digite seu primeiro numero:"))
c2=float(input("Digite seu segundo numero:"))
c3=float(input("Digite seu terceiro numero:"))

if c1>c2 and c1>c3:
    print("O primeiro numero é o maior")
elif c2>c1 and c2>c3:
    print("O segundo numero é o maior")
else:
    print("O terceiro numero e o maior")
```

10. Sistema simples de login

```python
# Sistema simples de login
usuario = input("Digite seu usuário: ")
senha = input("Digite sua senha: ")

usuario_correto = "admin"
senha_correta = "1234"  # ← String, para comparar com o input()

# Verifica se AMBOS estão corretos
if usuario == usuario_correto and senha == senha_correta:
    print("Login realizado com sucesso")
else:
    print("Usuário ou senha incorretos")
```
