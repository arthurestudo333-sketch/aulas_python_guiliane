```python
#Verificação de idade
idade = int(input("Digite sua idade:"))

if idade >=18:
    print("Maior de idade")

else:
    print("Menor de idade")
```

```python
#Calculador de media
nota = int(input("Digite sua nota"))

if nota >=7:
    print("Aprovado")

elif nota >=5:
    print("Recuperação")

else:
    print("Reprovado")
```

```python
#Verificando "==" e "!="
num_digitado = float(input("Digite um numero entre -2 e 5"))
num_digitado

if num_digitado == -1:
    print("Você acertou!")
else:
    print("Que pena, você errou!")
```

```python
if num_digitado != 3:
    print("Que pena, tente novamente")
else:
    print("Você acertou!")
```

```python
#Resto da divisão por um numero '%'
#par ou impar
num = int(input("Digite um numero natural:"))

if num%2 == 0:
    print(f"O {num} é par")
else:
    print(f" O {num} é impar!")
```

```python
#Comparar 2 números
num1 = float(input("Digite o primeiro numero:"))
num2 = float(input("Digite o segundo numero:"))

if num1 > num2:
    print(f"{num1} > {num2}")
    
elif num1 == num2:
    print(f"{num1} = {num2}")  
    
else:
    print(f"{num1} < {num2}")
```

```python
#Equação segundo grau
a= float(input("Digite o valor do coeficiente a"))
b= float(input("Digite o valor do coeficiente b"))
c= float(input("Digite o valor do coeficiente c"))

delta = b**2 - 4*a*c
x1= (- b + delta**(0.5))/(2*a)
x2= (- b - delta**(0.5))/(2*a)

if delta < 0:
    print(f"Delta menor que zero {delta}")
    print("Não há raiz real")
elif a == 0:
    print(f"{a} é menor que zero, insira um número diferente de zero")
elif delta == 0:
    print(f" x'= x''={x2}")
else:
    print(f"x'={x1} e x''={x2}")
```
